```python
from scipy.io import loadmat
#十进制浮点数转二进制
def dectbin(num):
    # 判断是否为浮点数
    if num == int(num):
        # 若为整数
        integer = '{:b}'.format(int(num))
        return integer
    else:
        # 若为浮点数
        # 取整数部分
        integer_part = int(num)
        # 取小数部分
        decimal_part = num - integer_part
        # 整数部分进制转换
        integercom = '{:b}'.format(integer_part)  #{:b}.foemat中b是二进制
        if len(integercom) < 8:
            a = []
            for i in range(8 - len(integercom)):
                a.append('0')
            integercom = ''.join(a) + integercom
        # 小数部分进制转换
        tem = decimal_part
        tmpflo = []
        # for i in range(accuracy):
        A = True
        while A:
            tem *= 2
            tmpflo += str(int(tem))  #若整数部分为0则二进制部分为0，若整数部分为1则二进制部分为1 #将1或0放入列表
            if tem > 1 :   #若乘以2后为大于1的小数，则要减去整数部分
                tem -= int(tem)
            elif tem < 1:  #乘以2后若仍为小于1的小数，则继续使用这个数乘2变换进制
                pass
            else:    #当乘以2后正好为1，则进制变换停止
                break
        flocom = []
        for i in range(8):
            flocom.append(tmpflo[i])
        return integercom + ''.join(flocom)
    
if __name__ == '__main__':
    file = './Desktop/X.mat'
    # mat_dtype=True，保证了导入后变量的数据类型与原类型一致。
    data = loadmat(file, mat_dtype=True)
    for i in range(167):
        number = data['X'][0][i] + 5.2
        result = dectbin(number)
        print(result)

    
```
